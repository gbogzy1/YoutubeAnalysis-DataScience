###  Data Preprocessing and Feature Generation

In order to utilize the data for analysis, several pre-processing steps needed to be undertaken. 
  - Preprocessing
      - I initially performed a search on the data types and also checked to see if we had any substantial null values that stood out.
  - Feature Generation 
      - It was necessary to reformat certain columns, particularly those related to date and time, such as "publishedAt" and "duration". I did a check  
      - Additionally, enriching the data with new features that can provide valuable insights into the characteristics of the videos would be quite essential i.e tagCount - to see how many tags are per video.

```python
#data preprocessing - looking for any column where there are missing(null) values
video_df.isnull().any()
```

```python
# assessing the data types
video_df.dtypes 
```

Realising that the channel data is in string format, I converted to numerical form

```python

# Making the count columns numerical as we can see they are strings.

numeric_cols = ['viewCount', 'likeCount', 'favouriteCount','commentCount']
video_df[numeric_cols] = video_df[numeric_cols].apply(pd.to_numeric, errors = 'coerce', axis =1)
```

```python

# Making a column for the day the video was posted to further enrich data.

video_df['publishedAt'] = video_df['publishedAt'].apply(lambda x: parser.parse(x))
video_df['publishedDayName'] = video_df['publishedAt'].apply(lambda x: x.strftime("%A"))
```

```python

# Adding a tag count - counting how many tags if any.
video_df['tagCount'] = video_df['tags'].apply(lambda x:0 if x is None else len(x))
```

###¬†Finding the top 10 videos (view count)

```python
ax = sns.barplot(x = 'title', y = 'viewCount', data = video_df.sort_values('viewCount',ascending=False)[0:10])
plot = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

Wondering what is causing the huge gap between top 5 - I will add some more metrics and compare further down

```python
video_df['likeRatio'] = video_df['likeCount']/ video_df['viewCount'] * 1000
video_df['commentRatio'] = video_df['commentCount']/ video_df['viewCount'] * 1000

#how long is the title - how many characters?
video_df['titleLength'] = video_df['title'].apply(lambda x: len(x))
```

## VIDEO DURATION

```python
# convert duration to seconds
video_df['durationSecs'] = video_df['duration'].apply(lambda x: isodate.parse_duration(x))
video_df['durationSecs'] = video_df['durationSecs'].astype('timedelta64[s]')
```

```python
sns.histplot(data = video_df, x = 'durationSecs', bins=30)
```

```python
fig, ax =plt.subplots(1,2)
sns.scatterplot(data = video_df, x = "durationSecs", y = "commentCount", ax=ax[0])
sns.scatterplot(data = video_df, x = "durationSecs", y = "likeCount", ax=ax[1])
```

We see that short videos under 400 seconds have virtually no comments and likes. Tope performers seem to be the mid 600 - 100s. But yet again, there is no direct relationship.

# Conclusions and Future Research Ideas

Throughout this project, we delved into the video data of my favourite youtuber, Obi Vincent.

- We observed a correlation between the number of likes and comments a video receives and its view count. However, it's important to note that this correlation does not necessarily imply causation. Likes appear to be a stronger indicator of viewer interaction compared to comments. Furthermore, the number of likes seems to follow the principle of "social proof," where videos with higher view counts tend to accumulate more likes.
- Most videos have a moderate number of tags, typically ranging between 20 and 40. This finding suggests that a balanced approach to utilizing tags can be beneficial for video visibility and discoverability.
